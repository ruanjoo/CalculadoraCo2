```python
anls = pd.read_csv("carros_dados.csv")


display(anls)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Ajustar o DataFrame x para incluir apenas as colunas relevantes
x = anls.iloc[:, [0, 1]]  # Usando os índices das colunas

# Extrair as colunas y
y_Conv_Un_Comum = anls.iloc[:, 2]  # Usando o índice da coluna Conv. Un. Comum
y_Cont_Carbono = anls.iloc[:, 3]  # Usando o índice da coluna Cont. Carbono
y_EmissCO2 = anls.iloc[:, 4]       # Usando o índice da coluna EmissCO2


# Separar os dados em treino e teste para Conv. Un. Comum
x_treino, x_teste, y_treino_Conv_Un_Comum, y_teste_Conv_Un_Comum = train_test_split(x, y_Conv_Un_Comum, test_size=0.3, random_state=1)

# Separar os dados em treino e teste para Cont. Carbono
x_treino, x_teste, y_treino_Cont_Carbono, y_teste_Cont_Carbono = train_test_split(x, y_Cont_Carbono, test_size=0.3, random_state=1)

# Separar os dados em treino e teste para EmissCO2
x_treino, x_teste, y_treino_EmissCO2, y_teste_EmissCO2 = train_test_split(x, y_EmissCO2, test_size=0.3, random_state=1)

# Instanciar o modelo de regressão linear para Conv. Un. Comum
modelo_Conv_Un_Comum = LinearRegression()
# Treinar o modelo para Conv. Un. Comum
modelo_Conv_Un_Comum.fit(x_treino, y_treino_Conv_Un_Comum)

# Instanciar o modelo de regressão linear para Cont. Carbono
modelo_Cont_Carbono = LinearRegression()
# Treinar o modelo para Cont. Carbono
modelo_Cont_Carbono.fit(x_treino, y_treino_Cont_Carbono)

# Instanciar o modelo de regressão linear para EmissCO2
modelo_EmissCO2 = LinearRegression()
# Treinar o modelo para EmissCO2
modelo_EmissCO2.fit(x_treino, y_treino_EmissCO2)
```

```python
from sklearn.metrics import mean_squared_error

# Fazer previsões nos dados de teste para Conv. Un. Comum
previsoes_Conv_Un_Comum = modelo_Conv_Un_Comum.predict(x_teste)
mse_Conv_Un_Comum = mean_squared_error(y_teste_Conv_Un_Comum, previsoes_Conv_Un_Comum)
print("MSE Conv. Un. Comum:", mse_Conv_Un_Comum)

# Fazer previsões nos dados de teste para Cont. Carbono
previsoes_Cont_Carbono = modelo_Cont_Carbono.predict(x_teste)
mse_Cont_Carbono = mean_squared_error(y_teste_Cont_Carbono, previsoes_Cont_Carbono)
print("MSE Cont. Carbono:", mse_Cont_Carbono)

# Fazer previsões nos dados de teste para EmissCO2
previsoes_EmissCO2 = modelo_EmissCO2.predict(x_teste)
mse_EmissCO2 = mean_squared_error(y_teste_EmissCO2, previsoes_EmissCO2)
print("MSE EmissCO2:", mse_EmissCO2)
```

```python
novosCarros = pd.read_csv("novosCarros.csv")
previsoes = modelo_EmissCO2.predict(novosCarros)
print(previsoes)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error

# Instanciar o modelo Ridge
modelo_ridge = Ridge()

# Separar os dados em treino e teste para Conv. Un. Comum
x_treino, x_teste, y_treino_Conv_Un_Comum, y_teste_Conv_Un_Comum = train_test_split(x, y_Conv_Un_Comum, test_size=0.3, random_state=1)

# Treinar modelo Ridge para Conv. Un. Comum
modelo_ridge_Conv_Un_Comum = Ridge(alpha=1.0)
modelo_ridge_Conv_Un_Comum.fit(x_treino, y_treino_Conv_Un_Comum)

# Treinar modelo Lasso para Conv. Un. Comum
modelo_lasso_Conv_Un_Comum = Lasso(alpha=1.0)
modelo_lasso_Conv_Un_Comum.fit(x_treino, y_treino_Conv_Un_Comum)

# Separar os dados em treino e teste para Cont. Carbono
x_treino, x_teste, y_treino_Cont_Carbono, y_teste_Cont_Carbono = train_test_split(x, y_Cont_Carbono, test_size=0.3, random_state=1)

# Treinar modelo Ridge para Cont. Carbono
modelo_ridge_Cont_Carbono = Ridge(alpha=1.0)
modelo_ridge_Cont_Carbono.fit(x_treino, y_treino_Cont_Carbono)

# Treinar modelo Lasso para Cont. Carbono
modelo_lasso_Cont_Carbono = Lasso(alpha=1.0)
modelo_lasso_Cont_Carbono.fit(x_treino, y_treino_Cont_Carbono)

# Separar os dados em treino e teste para EmissCO2
x_treino, x_teste, y_treino_EmissCO2, y_teste_EmissCO2 = train_test_split(x, y_EmissCO2, test_size=0.3, random_state=1)

# Treinar modelo Ridge para EmissCO2
modelo_ridge_EmissCO2 = Ridge(alpha=1.0)
modelo_ridge_EmissCO2.fit(x_treino, y_treino_EmissCO2)

# Treinar modelo Lasso para EmissCO2
modelo_lasso_EmissCO2 = Lasso(alpha=1.0)
modelo_lasso_EmissCO2.fit(x_treino, y_treino_EmissCO2)
```

```python
# Fazer previsões nos dados de teste para Conv. Un. Comum com Ridge
previsoes_Conv_Un_Comum_ridge = modelo_ridge_Conv_Un_Comum.predict(x_teste)
mse_Conv_Un_Comum_ridge = mean_squared_error(y_teste_Conv_Un_Comum, previsoes_Conv_Un_Comum_ridge)
print("MSE Conv. Un. Comum (Ridge):", mse_Conv_Un_Comum_ridge)

# Fazer previsões nos dados de teste para Conv. Un. Comum com Lasso
previsoes_Conv_Un_Comum_lasso = modelo_lasso_Conv_Un_Comum.predict(x_teste)
mse_Conv_Un_Comum_lasso = mean_squared_error(y_teste_Conv_Un_Comum, previsoes_Conv_Un_Comum_lasso)
print("MSE Conv. Un. Comum (Lasso):", mse_Conv_Un_Comum_lasso)

# Fazer previsões nos dados de teste para Cont. Carbono com Ridge
previsoes_Cont_Carbono_ridge = modelo_ridge_Cont_Carbono.predict(x_teste)
mse_Cont_Carbono_ridge = mean_squared_error(y_teste_Cont_Carbono, previsoes_Cont_Carbono_ridge)
print("MSE Cont. Carbono (Ridge):", mse_Cont_Carbono_ridge)

# Fazer previsões nos dados de teste para Cont. Carbono com Lasso
previsoes_Cont_Carbono_lasso = modelo_lasso_Cont_Carbono.predict(x_teste)
mse_Cont_Carbono_lasso = mean_squared_error(y_teste_Cont_Carbono, previsoes_Cont_Carbono_lasso)
print("MSE Cont. Carbono (Lasso):", mse_Cont_Carbono_lasso)

# Fazer previsões nos dados de teste para EmissCO2 com Ridge
previsoes_EmissCO2_ridge = modelo_ridge_EmissCO2.predict(x_teste)
mse_EmissCO2_ridge = mean_squared_error(y_teste_EmissCO2, previsoes_EmissCO2_ridge)
print("MSE EmissCO2 (Ridge):", mse_EmissCO2_ridge)

# Fazer previsões nos dados de teste para EmissCO2 com Lasso
previsoes_EmissCO2_lasso = modelo_lasso_EmissCO2.predict(x_teste)
mse_EmissCO2_lasso = mean_squared_error(y_teste_EmissCO2, previsoes_EmissCO2_lasso)
print("MSE EmissCO2 (Lasso):", mse_EmissCO2_lasso)
```

```python
from sklearn.neural_network import MLPRegressor


# Separar os dados em treino e teste
x_treino, x_teste, y_treino_Conv_Un_Comum, y_teste_Conv_Un_Comum = train_test_split(x, y_Conv_Un_Comum, test_size=0.3, random_state=1)
_, _, y_treino_Cont_Carbono, y_teste_Cont_Carbono = train_test_split(x, y_Cont_Carbono, test_size=0.3, random_state=1)
_, _, y_treino_EmissCO2, y_teste_EmissCO2 = train_test_split(x, y_EmissCO2, test_size=0.3, random_state=1)

# Criar e treinar o modelo MLPRegressor
modelo_Conv_Un_Comum = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=1000, random_state=1)
modelo_Conv_Un_Comum.fit(x_treino, y_treino_Conv_Un_Comum)

modelo_Cont_Carbono = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=1000, random_state=1)
modelo_Cont_Carbono.fit(x_treino, y_treino_Cont_Carbono)

modelo_EmissCO2 = MLPRegressor(hidden_layer_sizes=(100,), activation='relu', solver='adam', max_iter=1000, random_state=1)
modelo_EmissCO2.fit(x_treino, y_treino_EmissCO2)

# Fazer previsões
previsoes_Conv_Un_Comum = modelo_Conv_Un_Comum.predict(x_teste)
previsoes_Cont_Carbono = modelo_Cont_Carbono.predict(x_teste)
previsoes_EmissCO2 = modelo_EmissCO2.predict(x_teste)

# Calcular os Mean Squared Errors (MSEs)
mse_Conv_Un_Comum = mean_squared_error(y_teste_Conv_Un_Comum, previsoes_Conv_Un_Comum)
mse_Cont_Carbono = mean_squared_error(y_teste_Cont_Carbono, previsoes_Cont_Carbono)
mse_EmissCO2 = mean_squared_error(y_teste_EmissCO2, previsoes_EmissCO2)

print("Mean Squared Error (Conv. Un. Comum):", mse_Conv_Un_Comum)
print("Mean Squared Error (Cont. Carbono):", mse_Cont_Carbono)
print("Mean Squared Error (EmissCO2):", mse_EmissCO2)
```
